```python
# 서울 부산 위도 경도 지정. 초기 zoom 10
seoul_map = folium.Map(location=[37.56667,126.97806], zoom_start=10)
busan_map = folium.Map(location=[35.18,129.08], zoom_start=10)
```

```python
# 서울 부산 구분하기 위해 locate 에 도, 위도 경도 자료 가져옴. 여기서 x는 경도이고 y가 위도임. 나중에 지도 마킹할 때 조심
locate = pd.DataFrame(train[['addr_do', 'point.x', 'point.y']])
#locate1 = pd.DataFrame(train[['point.x', 'point.y']])
# point.x 로 인해 판다스 사용할 때 어려워서 이름 변경
locate.rename(columns={'point.x': 'pointx', 'point.y': 'pointy'}, inplace=True)
```

```python
# 서울, 부산 자료 빼기위해 불린 인덱싱 사용. isin 사용해도 뭐...
# 다음에는 필터 함수를..
se1 = locate['addr_do'] == "서울"
bu1 = locate['addr_do'] == "부산"
locate_se = locate.loc[se1, :]
locate_bu = locate.loc[bu1, :]
# 개수 맞는지 확인
len(locate) == len(locate_se) + len(locate_bu)
```

```python
# 서울맵
for lat, lng in zip(locate_se.pointy, locate_se.pointx):
    folium.Marker([lat, lng]).add_to(seoul_map)

#부산맵
for lat, lng in zip(locate_bu.pointy, locate_bu.pointx):
    folium.Marker([lat, lng]).add_to(busan_map)
```
